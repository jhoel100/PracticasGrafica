```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("1.png",cv2.IMREAD_GRAYSCALE)
#img = cv2.imread("1.png")
cv2.imshow('Inicial', img)

# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
#vuelve negro lo del intervalo
def Threshold(imagen,inf,sup):
    modificado=imagen
    #print(imagen)
    for i in range(len(imagen)):
        for j in range(len(imagen[i])):
            if(imagen[i][j] > inf and imagen[i][j] < sup):
                modificado[i][j] =np.uint8(0)
            else:
                modificado[i][j] =np.uint8(255)
    return modificado

#vuelve negro lo menor al valor
def Threshold2(imagen,sup):
    modificado=imagen
    #print(imagen)
    for i in range(len(imagen)):
        for j in range(len(imagen[i])):
            if(imagen[i][j] < sup):
                modificado[i][j] =np.uint8(0)
            else:
                modificado[i][j] =np.uint8(255)
    return modificado
```

```python
img = cv2.imread("1.png",cv2.IMREAD_GRAYSCALE)
f, (ax1,ax2) =plt.subplots(1, 2,sharey=True,figsize=(15, 6))
ax1.hist(img.ravel(),256,[0,256])

title_= "Imagen Original " +str(1)
ax1.set_title(title_)
ax1.set_xlabel('Intensidad')
ax1.set_ylabel('Cantidad de pixeles')
ax1.set_ylim(0,15000)

cv2.imshow(title_,img)

img2=Threshold2(img,184)

ax2.hist(img.ravel(),256,[0,256])
title_= "Histograma Maximo Valor "
ax1.set_title(title_)
ax2.set_xlabel('Intensidad')
ax2.set_ylabel('Cantidad de pixeles')
cv2.imshow(title_,img2)

# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
#otros ejemplos
import cv2
import numpy as np

image1 = cv2.imread('cheems.jpg')

img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

# all pixels value above 120 will be set to 255
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)

cv2.imshow('Binary Threshold', thresh1)
cv2.imshow('Binary Threshold Inverted', thresh2)
cv2.imshow('Truncated Threshold', thresh3)
cv2.imshow('Set to 0', thresh4)
cv2.imshow('Set to 0 Inverted', thresh5)

# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```
